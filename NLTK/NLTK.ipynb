### Find the 50 highest frequency word in Wall Street Journal corpus in NLTK.books (text7), All punctuation removed and all words lowercased.

```python
import nltk
from nltk.book import text7
```

```python
wsj = text7
```

Remove all the punctuation from the Wall Street Journal corpus and convert all the words to lowercase. This is done to ensure that each word is treated as a separate token for analysis and to reduce the total number of unique words in the corpus.

```python
# Step 2: Remove punctuation and convert to lowercase
wsj_words = [word.lower() for word in wsj if word.isalpha()]
```

Tokenize the Wall Street Journal corpus to get a list of individual words. Tokenization is the process of breaking down a text into individual words or tokens, which can then be analyzed further.

```python
# Step 3: Tokenize the corpus
wsj_tokens = nltk.FreqDist(wsj_words)
```

The wsj_tokens object is a FreqDist object that contains the frequency distribution of each word in the Wall Street Journal corpus. We convert this object to a list of tuples. We then sort this list of tuples based on the frequency count of each token. 

Finally, we set sort the list in descending order, so that the most frequent tokens appear first.

The resulting sorted_tokens list contains all the tokens in the Wall Street Journal corpus sorted by their frequency in descending order.

```python
# Step 4: Sort the words by frequency
sorted_tokens = sorted(wsj_tokens.items(), key=lambda x: x[1], reverse=True)
```

Extract only the first 50 (i.e. top 50 as list is in descending order)

```python
# Step 5: Get the top 50 most frequent words
top_50_tokens = sorted_tokens[:50]
```

```python
# Output only the first element, i.e. tokens from the tuple top_50_tokens
print([token[0] for token in top_50_tokens])
```

```python
# Step 6: Plot the Top 50 Tokens

# We set cumulative=False to plot a regular frequency distribution instead of a cumulative distribution
wsj_tokens.plot(50, cumulative=False)
```
